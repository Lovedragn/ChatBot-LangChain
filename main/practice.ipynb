```python
from langchain_google_genai import ChatGoogleGenerativeAI
from dotenv import load_dotenv

load_dotenv()

llm_model = ChatGoogleGenerativeAI(model="gemini-2.5-flash",temperature=0.8)
```

```python
from langchain.prompts import ChatPromptTemplate
from langchain.chains import LLMChain
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnablePassthrough
Chain1 = ChatPromptTemplate.from_messages([("human","tell me about {user_input} in very short")])
Chain2 = ChatPromptTemplate.from_messages([("human","no of words in the sentence : {word_input}")])

chain_prompt_1 = LLMChain(llm = llm_model , prompt=Chain1,output_key="word_input") 
chain_prompt_2 = LLMChain(llm = llm_model , prompt=Chain2,output_key="final_output")

parser = StrOutputParser()

chain =  chain_prompt_1 | chain_prompt_2
response =parser.parse(chain.invoke({"user_input":"ronaldo"}))
```
